Caja original. Extrae 2 canicas sin reemplazo. Dado que la primera fue roja, estime P(la segunda también sea roja)

```python
# Importar numpy para usar generación aleatoria y operaciones numéricas.
import numpy as np
# Fijar semilla para reproducibilidad del experimento.
np.random.seed(2026)
# Definir cantidad de repeticiones del experimento.
n_repeticiones = 10_000

caja_unica = np.array(['roja'] * 5 + ['azul'] * 3 + ['verde'] * 2)  # 10 canicas
# Preparar array booleano para marcar si en cada repetición salieron dos rojas.
ambas_rojas = np.zeros(n_repeticiones, dtype=bool)
primera_roja= np.zeros(n_repeticiones, dtype=bool)  # inicializar

for i in range(n_repeticiones):
    # Extraer dos canicas sin reemplazo de la caja.
    extraccion = np.random.choice(caja_unica, size=2, replace=False)
    # Marcar si ambas son rojas.
    ambas_rojas[i] = np.all(extraccion == 'roja')
    primera_roja[i] = extraccion[0] == 'roja'

probabilidad_dos_rojas_dada_primera_roja = np.mean(ambas_rojas)/np.mean(primera_roja)

print(f"P(Ambas rojas dado que la primera fue roja): {probabilidad_dos_rojas_dada_primera_roja:.4f}")
```

Lanza dos dados justos. Estima P(la suma es un número primo)

```python
# Importar numpy para usar generación aleatoria y operaciones numéricas.
import numpy as np
# Fijar semilla para reproducibilidad del experimento.
np.random.seed(2026)
# Definir cantidad de repeticiones del experimento.
n_repeticiones = 10000

# ============================================================
# Problema A - Dados
# ============================================================
# Generar n_repeticiones valores enteros uniformes 1..6 para el primer dado.
dado1 = np.random.randint(1, 7, size=n_repeticiones)  # primer dado
# Generar n_repeticiones valores enteros 1..6 para el segundo dado.
dado2 = np.random.randint(1, 7, size=n_repeticiones)  # segundo dado
# Sumar ambos resultados para cada repetición.
suma_dados = dado1 + dado2  # suma entre 2 y 12
# Crear indicador de suma igual a 7 por repetición.
suma_primo = (suma_dados == 2 ) | (suma_dados == 3 ) | (suma_dados == 5 ) | (suma_dados == 7 ) | (suma_dados == 11 )
prob_suma_primo = np.mean(suma_primo)  
print('Probabilidad de que la suma sea un primo')
print(prob_suma_primo)
```

Caja original. Extrae 3 canicas sin reemplazo. Dado que se obtuvo al menos una verde, estime P(exactamente dos rojas)

```python
# Importar numpy para usar generación aleatoria y operaciones numéricas.
import numpy as np
# Fijar semilla para reproducibilidad del experimento.
np.random.seed(2026)
# Definir cantidad de repeticiones del experimento.
n_repeticiones = 10000

caja_unica = np.array(['roja'] * 5 + ['azul'] * 3 + ['verde'] * 2)  # 10 canicas
# Preparar array booleano para marcar si en cada repetición salieron dos rojas.
dos_rojas_una_verde = np.zeros(n_repeticiones, dtype=bool)
al_menos_una_verde= np.zeros(n_repeticiones, dtype=bool)  # inicializar
for i in range(n_repeticiones):
    # Extraer dos canicas sin reemplazo de la caja.
    extraccion = np.random.choice(caja_unica, size=3, replace=False)
    # Marcar si ambas son rojas.
    dos_rojas_una_verde[i] = (extraccion[0] == 'verde' and extraccion[1] == 'roja' and extraccion[2] == 'roja') or (extraccion[1] == 'verde' and extraccion[0] == 'roja' and extraccion[2] == 'roja') or (extraccion[2] == 'verde' and extraccion[1] == 'roja' and extraccion[0] == 'roja')
    al_menos_una_verde[i] = np.any(extraccion == 'verde')

print(al_menos_una_verde)
print(dos_rojas_una_verde)

respuesta = np.mean(dos_rojas_una_verde)/np.mean(al_menos_una_verde)
print("Probabilidad de que se obtengan exactamente dos rojas dado que se obtuvo al menos una verde")
print(f"{respuesta:.4f}")
```
